```python
#Import the libraries
import math
import pandas_datareader as pdr
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("Nifty 500.xlsx")
```

```python
df = df[df['Volume']>500000]
df.head()
```

```python
stocks = pd.DataFrame()
for i in df["Ticker"]:
    try:
        stock = pdr.DataReader(i+".NS", data_source='yahoo', start='2020-08-12', end='2020-08-12')
        stock['Stock'] = i
        stocks = stocks.append(stock)
    except:
        print(i+": No Data")
        pass
stocks.set_index('Stock', inplace=True)
```

```python
stocks.to_excel("stocks1208.xlsx")
```
