```python
df_ZEEL = pd.read_excel("Zeel2002to2020.xlsx", index_col=0, parse_dates=True)
df_ZEEL.head()
```

```python
plt.figure(figsize=(12.2, 4.5))
plt.plot(df_ZEEL['Close'], label = 'Close Price')
plt.title('ZEEL Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend(df_ZEEL.columns.values, loc='upper left')
plt.show()
```

```python
df_ZEEL['Buy'] = np.nan
```

```python
for i in range(4, df_ZEEL.shape[0]):
    if df_ZEEL.iloc[i-1]['Open'] > df_ZEEL.iloc[i-1]['Close'] and \
       df_ZEEL.iloc[i-2]['Open'] > df_ZEEL.iloc[i-2]['Close'] and \
       df_ZEEL.iloc[i-3]['Open'] > df_ZEEL.iloc[i-3]['Close'] and \
       df_ZEEL.iloc[i-4]['Open'] > df_ZEEL.iloc[i-4]['Close'] and \
       df_ZEEL.iloc[i]['Close'] > df_ZEEL.iloc[i]['Open']:
        print(df_ZEEL.index[i].date())
        df_ZEEL['Buy'].iloc[i] = df_ZEEL['Close'].iloc[i]
        print()
```
