```python
#Import the libraries
# import math
import pandas_datareader as pdr
import datetime
# import mplfinance as mpf
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("Nifty 500.xlsx")
df = df[df['Volume']>500000]
tickers = df.Ticker.tolist()
tickers = [x+".NS" for x in tickers]
```

```python
aWeekAgo = (datetime.datetime.today() - datetime.timedelta(days=7)).strftime("%Y-%m-%d")
today = datetime.datetime.today().strftime('%Y-%m-%d')
stocks = pdr.DataReader(tickers, data_source='yahoo', start=aWeekAgo, end=today)
```

```python
count = 0
values = []
for i in tickers:
    if stocks[-1:]["Open"][i][0] == stocks[-1:]["Low"][i][0]:
        count += 1
        values.append(i.split(".")[0])
        
print(count)
df.loc[df[df['Ticker'].isin(values)].index][['Ticker','Volume']].sort_values(["Volume"],ascending=False)
```

```python
def open_equalsto_low(df):
    # Testing Open == Low Strategy
    count = 0
    count2 = 0
    for index, row in df[:-1].iterrows():
        if row['Low'] == row['Open']:
            count += 1
#             print(f"Open and Close Same day: {row['Open']}, {row['Close']}")
#             print(f"Open, Close and High next day: {df['Open'].iloc[index+1]}, {df['Close'].iloc[index+1]}, {df['High'].iloc[index+1]}")
            if(df['Open'].iloc[index+1] < df['High'].iloc[index+1] and (((df['High'].iloc[index+1] - df['Open'].iloc[index+1])/df['High'].iloc[index+1]) * 100)>1):
                count2 += 1
#             print(((df['High'].iloc[index+1] - df['Open'].iloc[index+1])/df['High'].iloc[index+1]) * 100)
#             print()
    return count,count2
    
```

```python
def open_equalsto_high(df):
    # Testing Open == High Strategy
    count = 0
    count2 = 0
    for index, row in df[:-1].iterrows():
        if row['High'] == row['Open']:
            count += 1
#             print(f"Open and Close Same day: {row['Open']}, {row['Close']}")
#             print(f"Open, Close and High next day: {df['Open'].iloc[index+1]}, {df['Close'].iloc[index+1]}, {df['Low'].iloc[index+1]}")
            if(df['Open'].iloc[index+1] > df['Low'].iloc[index+1] and (((df['Open'].iloc[index+1] - df['Low'].iloc[index+1])/df['Open'].iloc[index+1]) * 100)>1):
                count2 += 1
#             print(((df['Open'].iloc[index+1] - df['Low'].iloc[index+1])/df['Open'].iloc[index+1]) * 100)
#             print()

    return count,count2
```
