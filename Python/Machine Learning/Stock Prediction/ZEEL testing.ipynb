```python
# Testing Open == Low Strategy
count = 0
count2 = 0
for index, row in df.iterrows():
    if row['Low'] == row['Open']:
        count += 1
        print(f"Open and Close Same day: {row['Open']}, {row['Close']}")
        print(f"Open, Close and High next day: {df['Open'].iloc[index+1]}, {df['Close'].iloc[index+1]}, {df['High'].iloc[index+1]}")
        if(df['Open'].iloc[index+1] < df['High'].iloc[index+1] and (((df['High'].iloc[index+1] - df['Open'].iloc[index+1])/df['High'].iloc[index+1]) * 100)>2):
            count2 += 1
        print(((df['High'].iloc[index+1] - df['Open'].iloc[index+1])/df['High'].iloc[index+1]) * 100)
        print()

print(f"Total case: {count}")
print(f"True case: {(count2)}")
print(f"Success Rate: {round((count2/count)*100)}%")
print()
```

```python
# Testing Open == High Strategy
count = 0
count2 = 0
for index, row in df.iterrows():
    if row['High'] == row['Open']:
        count += 1
        print(f"Open and Close Same day: {row['Open']}, {row['Close']}")
        print(f"Open, Close and High next day: {df['Open'].iloc[index+1]}, {df['Close'].iloc[index+1]}, {df['Low'].iloc[index+1]}")
        if(df['Open'].iloc[index+1] > df['Low'].iloc[index+1] and (((df['Open'].iloc[index+1] - df['Low'].iloc[index+1])/df['Open'].iloc[index+1]) * 100)>1):
            count2 += 1
        print(((df['Open'].iloc[index+1] - df['Low'].iloc[index+1])/df['Open'].iloc[index+1]) * 100)
        print()

print(f"Total case: {count}")
print(f"True case: {(count2)}")
print(f"Success Rate: {round((count2/count)*100)}%")
print()
```
