```python
col_names =  ['High', 'Low', 'Close', 'Close Date'] 
df_new  = pd.DataFrame(columns = col_names) 
# df_new  = pd.DataFrame({'Week': ['na'], 'High': ['na'], 'Low': ['na'], 'Close':['na']})  
```

```python
for i in range(df_ZEEL.shape[0]):
    if df_ZEEL.index[i].weekday() == 0:
        try:
            df_new.loc[len(df_new.index)] = [df_ZEEL[df_ZEEL.index[i].date():df_ZEEL.index[i+4].date()]['High'].max(), df_ZEEL[df_ZEEL.index[i].date():df_ZEEL.index[i+4].date()]['Low'].min(), df_ZEEL.iloc[i+4]['Close'], df_ZEEL.index[i+4].date()]
        except:
            print(df_ZEEL.index[i+3].date())
```

```python
df_new.index += 1
df_new.index.name = 'Week'
df_new['Call'] = 'na'
```
